### 1. 動態網頁跟靜態網頁的差別是什麼？原本靜態網頁的做法會產生什麼問題或是缺點？

- 靜態網⾴：靜態網頁的請求與回傳是⼀對⼀的，每⼀次使⽤者發出請求，後端會產⽣⼀次網⾴回傳，使用者的操作都由伺服器做出相對應的固定回應，沒有互動式的設計或讀取資料庫的過程，結構較動態網頁簡單，有些⼈把他們稱為同步。

- 動態網頁：動態網⾴透過 Ajax 的技術完成非同步的資料傳輸。當使用者送出不同請求時，不需要像靜態網頁一樣回傳整個葉面，而是依照使用者發出的請求動態產生相對應內容，也就是在網⾴上，任何時間點都可以發送請求給後端，後端只回傳資料至目前的頁面中，常見的應用如：留言板、表情符號等。

- 靜態網頁的問題與缺點：
    - 由於每次使用者發出一項請求時，靜態網頁都要重新載入新頁面，例如每點擊一次頁面上的按鈕，因為重新載入檔案的緣故，頁面就會閃一下，使用者體驗很差
    - 相較於動態網頁較不活潑
    - 規模較大的頁面難以維護，需要直接修改HTML及CSS內容

### 2. 還記得我們在 ETTODAY 靜態爬蟲實作練習中，有請大家完成一個題目「取出今天所有的發文」，但仔細看其實並沒有真的把當天所有的新聞都抓回來， 試著回答看看原因是什麼？及該如何檢查？

ETTODAY新聞內容會隨著使用者的捲動載入，如同臉書或 Instagram一樣，在向下滑的時候會不斷載入新內容。在之前的練習中之所以爬取的新聞不完整，是因為只透過 Python送出一次 URL以取得剛載入的網頁內容，並沒有執行捲動的行為，因此不會載入更多新聞。
若想要檢查是否完整爬取新聞，可以直接開啟瀏覽器觀看該網站是否有動態載入的設計。
