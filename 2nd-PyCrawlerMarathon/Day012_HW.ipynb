### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, 'html5lib')

news_dic = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    catalog = d.find(class_="tag").text
    title = d.find('a').text
    
    if catalog not in news_dic.keys():
        news_dic.update({catalog : []})
   
    news_dic[catalog].append(title)
    
news_dic
```
