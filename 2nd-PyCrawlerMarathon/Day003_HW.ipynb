### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


- File I/O讀取XML檔案，得到的是「str」類別
- xmltodict讀取XML檔案，得到的是「collections.OrderedDict」類別，再經由`dict`方法處理後會變成`dictionary`

```python
# File I/O
# 讀檔案
fh = open("./Day003/64_72hr_CH.xml", "r", encoding='utf8')
xml = fh.read()
fh.close()

# print(xml)
print(type(xml))
```

```python
# xmltodict
# 解析檔案內容
import xmltodict

temp = xmltodict.parse(xml)
print(type(temp))

d = dict(xmltodict.parse(xml))
print(type(d))

# 取出 datasetDescription
# datasetDescription = d['cwbopendata']['dataset']['datasetInfo']['datasetDescription']
# print(datasetDescription)
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

# 讀檔案
fh = open("./Day003/64_72hr_CH.xml", "r", encoding='utf8')
xml = fh.read()
fh.close()

# 解析檔案內容
import xmltodict
d = dict(xmltodict.parse(xml))

# 擷取各行政區資料
location = d['cwbopendata']['dataset']['locations']['location']

# 計數器
count = 0

# 以行政區數量跑迴圈
for i in range (0,len(location)):
    # 溫度資料
    weatherElement = d['cwbopendata']['dataset']['locations']['location'][i]['weatherElement']
    
    # 檢查每一筆溫度資料內是否有內容    
    if weatherElement:
        count += 1

print("共"+ str(count) + "個地區有溫度資料")
```

```python

# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for i in range (0,len(location)):
    # 區名稱
    locationName = d['cwbopendata']['dataset']['locations']['location'][i]['locationName']
    time = d['cwbopendata']['dataset']['locations']['location'][i]['weatherElement'][0]['time'][0]['dataTime']
    temperture = d['cwbopendata']['dataset']['locations']['location'][i]['weatherElement'][0]['time'][0]['elementValue']['value']
    
    print("行政區名稱："+locationName)
    print("時間："+ time)
    print("溫度："+ temperture +"(攝氏度)")
    print()
```
