## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
non_r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')

responses = json.loads(r.text)
non_responses = json.loads(non_r.text)

print("一共回傳 "+ str(len(responses)) +" 筆資料")
```

### 欄位資料如下圖



```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for data in responses:
    print(data['title'])
    print(data['createdAt'])
    print(data['commentCount'])
    print(data['likeCount'])
    print()
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

sum_comment = 0
sum_like = 0

for data in responses:
    sum_comment += data['commentCount']
    sum_like += data['likeCount']

print("熱門文章")
print("平均留言人數: "+ str(sum_comment/len(responses)))
print("平均按讚人數: "+ str(sum_like/len(responses)))
      
sum_comment = 0
sum_like = 0
    
for data in non_responses:
    sum_comment += data['commentCount']
    sum_like += data['likeCount']

print()
    
print("非熱門文章")
print("平均留言人數: "+ str(sum_comment/len(non_responses)))
print("平均按讚人數: "+ str(sum_like/len(non_responses)))
```
