```python
# 1. ÈÄôÂÄã API ‰∏ÄÊ¨°ÊúÉÂõûÂÇ≥ÂπæÁ≠ÜË≥áÊñôÔºüÊØè‰∏ÄÁ≠ÜË≥áÊñôÂåÖÂê´Âì™‰∫õÊ¨Ñ‰ΩçÔºü

import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
non_r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')

responses = json.loads(r.text)
non_responses = json.loads(non_r.text)

print("‰∏ÄÂÖ±ÂõûÂÇ≥ "+ str(len(responses)) +" Á≠ÜË≥áÊñô")
```

### Ê¨Ñ‰ΩçË≥áÊñôÂ¶Ç‰∏ãÂúñ



```python
# 2. ÂèñÂá∫ÊØè‰∏ÄÁ≠ÜË≥áÊñôÁöÑ„ÄåÊ®ôÈ°å„Äç„ÄÅ„ÄåË≤ºÊñáÊôÇÈñì„Äç„ÄÅ„ÄåÁïôË®Ä‰∫∫Êï∏„Äç„ÄÅ„ÄåÊåâËÆö‰∫∫Êï∏„Äç

for data in responses:
    print(data['title'])
    print(data['createdAt'])
    print(data['commentCount'])
    print(data['likeCount'])
    print()
```

```python
# 3. Ë®àÁÆóÁÜ±ÈñÄ/ÈùûÁÜ±ÈñÄÊñáÁ´†ÁöÑ„ÄåÂπ≥ÂùáÁïôË®Ä‰∫∫Êï∏„ÄçËàá„ÄåÂπ≥ÂùáÊåâËÆö‰∫∫Êï∏„Äç

sum_comment = 0
sum_like = 0

for data in responses:
    sum_comment += data['commentCount']
    sum_like += data['likeCount']

print("ÁÜ±ÈñÄÊñáÁ´†")
print("Âπ≥ÂùáÁïôË®Ä‰∫∫Êï∏: "+ str(sum_comment/len(responses)))
print("Âπ≥ÂùáÊåâËÆö‰∫∫Êï∏: "+ str(sum_like/len(responses)))
      
sum_comment = 0
sum_like = 0
    
for data in non_responses:
    sum_comment += data['commentCount']
    sum_like += data['likeCount']

print()
    
print("ÈùûÁÜ±ÈñÄÊñáÁ´†")
print("Âπ≥ÂùáÁïôË®Ä‰∫∫Êï∏: "+ str(sum_comment/len(non_responses)))
print("Âπ≥ÂùáÊåâËÆö‰∫∫Êï∏: "+ str(sum_like/len(non_responses)))
```
