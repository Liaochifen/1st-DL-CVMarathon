## 1. 取出今天所有的新聞

```python
from bs4 import BeautifulSoup 
from selenium import webdriver
from datetime import datetime
import time
```

```python
datetime.now().day

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    article_date = int(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text[8:10])
    if (datetime.now().day - article_date) >= 1:
        print('==== STOP ====')
        break
```

```python
html = browser.page_source
soup = BeautifulSoup(html, 'html5lib')

for d in soup.find(class_="part_list_2").find_all('h3'):
    article_date = int(d.find(class_="date").text[8:10])
    if (datetime.now().day - article_date) >= 1:
        break
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
# Close browser
browser.close()
```

## 2. 取出現在時間兩小時內的新聞

```python
from bs4 import BeautifulSoup 
from selenium import webdriver
from datetime import datetime
```

```python
time = datetime.now()
current_time = time.hour*60 + time.minute

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

print('共 %s 筆' %len(soup.find(class_="part_list_2").find_all('h3')))
for d in soup.find(class_="part_list_2").find_all('h3'):
    news_time = (int(d.find(class_="date").text[11:13])*60) + int(d.find(class_="date").text[14:17])
    if (current_time - news_time) <= 120:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        
# Close browser
browser.close()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from bs4 import BeautifulSoup 
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from datetime import datetime
import requests
import time

```

```python
current_time = datetime.now()

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

selectSite = Select(browser.find_element_by_id("selD"))
selectSite.select_by_value(str(current_time.day-3))

browser.find_element_by_id('button').click()

while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    article_hour = int(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text[11:13])
    if article_hour < 15:
        print('==== STOP ====')
        break
    
```

```python
html = browser.page_source
soup = BeautifulSoup(html, "html5lib")

time_range = ['15', '16', '17']

for d in soup.find(class_="part_list_2").find_all('h3'):
    hour = d.find(class_="date").text[11:13]
    minute = d.find(class_="date").text[14:17]
    
    if hour in time_range:
        if hour == time_range[2] and int(minute) <= 0:
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
        elif hour != time_range[2]:
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
            
# Close browser
browser.close()
```
