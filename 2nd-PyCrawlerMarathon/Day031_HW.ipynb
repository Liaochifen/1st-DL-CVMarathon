### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）
- Ans: 以 amazon.com 為例，送出請求時必須在 header 加上user-agent才能通過檢查

```python
import requests
r = requests.get('https://www.amazon.com/')
response = r.text

response
```

### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
import requests
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
r = requests.get('https://www.amazon.com/',headers=headers)
response = r.text

response
```
