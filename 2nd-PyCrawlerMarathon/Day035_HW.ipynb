## 作業目標

* 找一個之前實作過的爬蟲改用多線程改寫，比較前後時間的差異。





```python
from bs4 import BeautifulSoup 
from selenium import webdriver
from datetime import datetime
import time
import _thread
```

```python
def scroll(url, startTime):
    browser = webdriver.Chrome(executable_path='chromedriver')
    browser.get(url)
    
    while True:
        browser.execute_script("window.scrollTo(0, 100000);")
        time.sleep(2)
        html_source = browser.page_source
        soup = BeautifulSoup(html_source, "html5lib")
        article_date = int(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text[8:10])
        if (datetime.now().day - article_date) >= 1:
            print('==== STOP ====')
            break
            
    parse_content(soup, browser, startTime)
```

```python
def parse_content(soup, browser, startTime):
    for d in soup.find(class_="part_list_2").find_all('h3'):
        article_date = int(d.find(class_="date").text[8:10])
        if (datetime.now().day - article_date) >= 1:
            break
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
    browser.close()
    
    finishTime = time.time()
    print(finishTime - startTime) # 利用 thread 的爬蟲所需時間
```

```python
urls = ['https://www.ettoday.net/news/news-list.htm', 'https://www.ettoday.net/news/news-list.htm', 'https://www.ettoday.net/news/news-list.htm']

startTime = time.time()

for url in urls:
    _thread.start_new_thread( scroll, (url, startTime) )
```
